## At least one hashtag

```python
def find_hashtag(sample):
    total_hashtag = []
    for entity in sample.loc[:, 'entities']:
        list_hashtags = entity.get('hashtags')
        hashtag = []
        for elem in list_hashtags:
            h = elem.get('text')
            hashtag.append(h)
        total_hashtag.append(hashtag)
    return total_hashtag
```

```python
sample1 = sample1.assign(hashtags=find_hashtag(sample1))
sample2 = sample2.assign(hashtags=find_hashtag(sample2))
```

```python
def one_hashtag(sample):
    total = 0
    for list_h in sample.loc[:, 'hashtags']:
        if len(list_h) > 0:
            total += 1
    return total
```

```python
one_hashtag1 = one_hashtag(sample1)
print('Number of Tweets in sample 1 that contain at least one hashtag:', one_hashtag1)
print('Percentage of Tweets in sample 1 that contain at least one hashtag:', one_hashtag1/len(sample1)*100, '%')
```

```python
one_hashtag2 = one_hashtag(sample2)
print('Number of Tweets in sample 2 that have at least one hashtag:', one_hashtag2)
print('Percentage of Tweets in sample 2 that have at least one hashtag:', one_hashtag2/len(sample2)*100, '%')
```

## Two hashtags or more

```python
def two_hashtag(sample):
    total = 0
    for list_h in sample.loc[:, 'hashtags']:
        if len(list_h) > 2:
            total += 1
    return total
```

```python
two_hashtag1 = two_hashtag(sample1)
print('Number of Tweets in sample 1 that contain two or more hashtags:', two_hashtag1)
print('Percentage of Tweets in sample 1 that conatin two or more hashtags:', two_hashtag1/len(sample1)*100, '%')
```

```python
two_hashtag2 = two_hashtag(sample2)
print('Number of Tweets in sample 2 that contain two or more hashtags:', two_hashtag2)
print('Percentage of Tweets in sample 2 that contain two or more hashtags:', two_hashtag2/len(sample2)*100, '%')
```

## Media

```python
def total_media(sample, numb):
    total_media = 0
    for entity in sample.loc[:, 'entities']:
        list_media = entity.get('media')
        if list_media is not None:
            total_media += 1
            
    
    print('Number of Tweets in sample', numb , 'that contain media:', total_media)
    print('Percentage of Tweets in sample', numb, 'that contain media:', total_media/len(sample)*100, '%')
```

```python
total_media(sample1, 1)
```

```python
total_media(sample2, 2)
```

## URLs

```python
def total_url(sample, numb):
    total_url = 0
    for entity in sample.loc[:, 'entities']:
        list_urls = entity.get('urls')
        if len(list_urls) > 0:
            total_url += 1
            
    
    print('Number of Tweets in sample', numb , 'that contain URLs:', total_url)
    print('Percentage of Tweets in sample', numb, 'that contain URLs:', total_url/len(sample)*100, '%')
```

```python
total_url(sample1, 1)
```

```python
total_url(sample2, 2)
```

## 50 most frequent hashtags

```python
hash1 = [val for sublist in find_hashtag(sample1) for val in sublist]
most_freq1 = pd.DataFrame(data=Counter(hash1).most_common(50), columns=['hashtag','occurences'])
most_freq1.index = np.arange(1, len(most_freq1)+1)
most_freq1
```

```python
hash2 = [val for sublist in find_hashtag(sample2) for val in sublist]
most_freq2 = pd.DataFrame(data=Counter(hash2).most_common(50), columns=['hashtag','occurences'])
most_freq2.index = np.arange(1, len(most_freq2)+1)
most_freq2
```

## 50 most frequent hashtag pairs

```python
def find_pair_hashtag(sample):
    pair_hashtag = []
    for list_hashtag in sample.loc[:, 'hashtags']:
        if len(list_hashtag) > 0:
            list_ = sorted(list_hashtag, key=str.lower)
            for i in range(0, len(list_)):
                for j in range(i+1, len(list_)):
                    pair_hashtag.append(tuple([list_[i], list_[j]]))
    return pair_hashtag
```

```python
pair1 = find_pair_hashtag(sample1)
most_freq_pair1 = pd.DataFrame(data=Counter(pair1).most_common(50), columns=['hashtag pairs','occurences'])
most_freq_pair1.index = np.arange(1, len(most_freq_pair1)+1)
most_freq_pair1
```

```python
pair2 = find_pair_hashtag(sample2)
most_freq_pair2 = pd.DataFrame(data=Counter(pair2).most_common(50), columns=['hashtag pairs','occurences'])
most_freq_pair2.index = np.arange(1, len(most_freq_pair2)+1)
most_freq_pair2
```

# Pairwise comparison

```python
def common_hashtags(sample1, sample2, top):
    intersection = 0
    same_hashtags = []
    for i in range(1, top+1):
        for j in range(1, top+1):
            if sample1['hashtag'][i] == sample2['hashtag'][j]:
                same_hashtags.append(sample1['hashtag'][i])
                intersection += 1
    union = 2*top - intersection
    ratio = intersection / union
    print('Number of common hashtags in top', top, ':', intersection)
    print('Ratio of common hashtags in top ', top, ':', ratio)
    print('Common hashtags :', same_hashtags)
```
